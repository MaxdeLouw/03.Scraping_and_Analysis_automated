```python
import pandas as pd
import sqlite3 as sqlite
from pathlib import Path
```

```python
supermarket_DF = pd.read_csv(r'..\Data\Processed\supermarket_product_data_clean.csv')
supermarket_DF.head()
```

```python
db_path = Path(r"../Data/Processed/supermarket_products.db")

conn = sqlite.connect(db_path)

supermarket_DF.to_sql(
    name="supermarket_products",
    con=conn,
    if_exists="replace",
    index=False,
)

conn.close()
```
