```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt
```

```python
jumbo_raw = pd.read_csv("../Data/Raw/jumbo_product_data.csv")
plus_raw = pd.read_csv("../Data/Raw/plus_product_data.csv")
```

```python
print(jumbo_raw.shape)
print(jumbo_raw.dtypes)
print(jumbo_raw.info())
```

```python
jumbo_clean = jumbo_raw.copy()
jumbo_clean['Date_Added'] = pd.to_datetime(jumbo_raw['Date_Added'].copy()).dt.date

jumbo_clean.head()
```

```python
#removing spaces from the product name
name_text = jumbo_clean['Product_Name'].iloc[0]
new_name = name_text.replace(' ','_')
print(new_name)

#obtaining product brand
Brand = jumbo_clean['Product_Name'].iloc[0].split('_')[0]
print(Brand)

#Extracting just the price
price_text = jumbo_clean['Product_Price'].iloc[0]
product_price = float(price_text
                    .split(' ')[1]
                    .split("\xa0")[1]
                    .replace(",", '.'))
print(product_price)

price_unit_text = jumbo_clean['Price_Per_Unit'].iloc[0]
unit_price = float(price_unit_text
                 .split(' ')[3]
                 .replace(",", '.'))
print(unit_price)

#calculating the weight of a product
weight_in_g = product_price / unit_price * 1000
print(weight_in_g)

#converting energy to kcal instead of kJ
energy_text = jumbo_clean['Energy_kj_100_g'].iloc[0]
energy_kJ = float(energy_text
                 .split(' ')[0])
energy_kcal = round((energy_kJ / 4.184) ,1)

print(energy_kcal)


nutrient_categories = jumbo_clean.columns[8:]

for nutrient in nutrient_categories:
    nutrient_name = nutrient.split("_")[0]
    multiple_of_100_g = float(weight_in_g / 100)
    # f"{nutrient_name}_In_Product" = int(jumbo_clean[nutrient].split(" ")[0]) * multiple_of_100_g
    nutrients_in_product = float(jumbo_clean[nutrient].iloc[0].split(" ")[0]) * multiple_of_100_g
    print(f"{nutrients_in_product} grams of {nutrient_name}")


```
